```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

#Importing the data
A = pd.read_excel('CANDY-HIERARCHY-2015-SURVEY-Responses.xlsx')
B = pd.read_excel('BOING-BOING-CANDY-HIERARCHY-2016-SURVEY-Responses.xlsx')
```

```python
#Viewing the head of the first data frame
A.head()
```

```python
#Viewing the head of the second data frame
B.head()
```

```python
#Chapter 7 manipulations (1 of 2)

#String manipulations
#Setting all string columns to uppercase for uniformity. Applying the capitalization only to the string columns.
for col in A.columns:
    for i in range(len(A)):
        if isinstance(A[col][i], str) == True:
            A[col][i] = A[col][i].upper()

for col in B.columns:
    for i in range(len(B)):
        if isinstance(B[col][i], str) == True:
            B[col][i] = B[col][i].upper()
```

```python
#Chapter 7 manipulations (2 of 2)

#Filling in missing data
#Filling all missing data with 'NOT ANSWERED' to allow for easier assessment of the survey results.
#Filling the missing data with 'JOY', 'DESPAIR', or 'MEH' could lead to biased results.
A = A.fillna('NOT ANSWERED')
B = B.fillna('NOT ANSWERED')
```

```python
#Chapter 8 manipulations (1 of 2)

#Merging the two data frames with an outer join to preserve the keys of both tables.
#Some questions in one table aren't answered in the other table, so those are filled with 'NOT ANSWERED' again.
C = pd.merge(A, B, how = 'outer')
C = C.fillna('NOT ANSWERED')

#Displaying the merged data frame so far
C
```

```python
#Chapter 8 manipulations (2 of 2)

#Reshaping the data by pivoting along the diagonal
C.stack().unstack(0)
```

```python
#Chapter 10 manipulations (1 of 2)

#Split-Apply-Combine
#I want to know what the last person said for those who are going trick-or-treating, so I split the data by that question and applied last() to find the last entry.        
C.groupby('Are you going actually going trick or treating yourself?').last()
```

```python
#Chapter 10 manipulations (2 of 2)

#Cross-tabulation
#I want to know what those who did or didn't go trick-or-treating think of Butterfinger bars, so I used cross-tabulation.
pd.crosstab(C['Are you going actually going trick or treating yourself?'], C[' [Butterfinger]'], margins = True)
```

```python
#Chapter 11 manipulations (1 of 2)

#Converting all timestamps to string format by iterating over the timestamps
from datetime import datetime
dates = C['Timestamp']
date_list = [] 

#Converting each date to a string, adding it to a list, and placing that list as the new timestamp column
for date in dates:
    date = date.strftime('%Y-%m-%d %H:%M:%S')
    date_list.append(date)

C['Timestamp'] = pd.Series(date_list)

#Confirming that the series is the correct type
C['Timestamp']
```

```python
from dateutil.parser import parse

#Converting the strings back into timestamps
C['Timestamp'] = pd.to_datetime(C['Timestamp'])

#Confirming that the series has returned to a datetime format
C['Timestamp']
```

```python
#Chapter 11 manipulations (2 of 2)

#Converting timestamps to periods using the same method as converting to strings
L = []
for dates in C['Timestamp']:
    dates = dates.to_period('T')
    L.append(dates)
    
C['Timestamp'] = pd.Series(L)

#Confirming that the timestamps have been converted to periods
C['Timestamp']
```

```python
#Converting the periods back into timestamps
ts_list = []
for dates in C['Timestamp']:
    dates = dates.to_timestamp()
    ts_list.append(dates)
    
C['Timestamp'] = pd.Series(ts_list)

#Confirming they have been returned to timestamps
C['Timestamp']
```
