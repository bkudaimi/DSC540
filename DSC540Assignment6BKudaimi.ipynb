```python
#Activity 11 

#1: Connecting to petsdb and checking whether the connection has been successful.
import sqlite3
import pandas as pd
try:
    conn = sqlite3.connect('petsdb')
    #Fetching the persons data
    Query1 = pd.read_sql('SELECT * FROM persons', conn)
    print('Connected.')
except:
    print('Not connected.')
```

```python
#2: Finding the different age groups in the persons database

#Getting the ages and number of people at each age in the table in order then using a for loop to iterate over them.
Query = conn.cursor().execute('SELECT count(*), age FROM persons GROUP BY age')
for people, ages in Query:
    print('There are {} people aged {}.'.format(people, ages))
```

```python
#3: Finding the age group with the most number of people

#Getting the ages and number of people at each age in the table in order by their person count
Query2 = conn.cursor().execute('SELECT count(*), age FROM persons GROUP BY age ORDER BY count(*)')

#Extracting the ages and the number of people at each age in order. 
#The last item in each list should be the age group with the most number of people.
people_list = []
age_list = []
for people, age in Query2:
    people_list.append(people)
    age_list.append(age)
    
print('The age with the most number of people is {} and there are {} people in it.'.format(age_list[len(age_list)-1], people_list[len(people_list)-1]))
```

```python
#4: Finding the people who do not have a last name
Query3 = pd.read_sql('SELECT * FROM persons WHERE last_name IS null', conn)
Query3
```

```python
#5: Finding how many people have more than one pet

#Counting the number of occurrences of each owner ID and filtering out the owner IDs who have 0 and 1 pets, then selecting those counts
owner_counts = conn.cursor().execute('SELECT count(*) FROM (SELECT count(owner_id) FROM pets GROUP BY owner_id HAVING count(owner_id) > 1)')

#Appending the number to a list to allow for printing
multiple_pets = []
for number in owner_counts:
    multiple_pets.append(number)
    
print('{} people have more than one pet.'.format(int(multiple_pets[0][0])))
```

```python
#6: Finding out how many pets have received treatment.

pets_treated = pd.read_sql('SELECT count(*) FROM pets WHERE treatment_done = 1', conn)['count(*)'][0]
print('{} pets have received treatment'.format(pets_treated))
```

```python
#7: Finding out how many pets have received treatment and the type of pet is known using a method similar to 6 above

pets_treated_known = pd.read_sql('SELECT count(*) FROM pets WHERE treatment_done = 1 AND pet_type IS NOT null', conn)['count(*)'][0]
print('{} pets of a known type have received treatment'.format(pets_treated_known))
```

```python
#8: Finding out how many pets are from the city of east port

#Joining the persons and pets tables by their ID, selecting the pets whose owner lives in east port, and counting them.
pets_east_port = pd.read_sql('SELECT count(*) FROM pets JOIN persons ON pets.owner_id = persons.id WHERE persons.city = "east port"', conn)
print('{} pets are from east port'.format(pets_east_port['count(*)'][0]))
```

```python
#9: Finding out how many pets are from the city of east port who received treatment using a method similar to 8 above

pets_east_port_treated = pd.read_sql('SELECT count(*) FROM pets JOIN persons ON pets.owner_id = persons.id WHERE persons.city = "east port" AND treatment_done = 1', conn)
print('{} treated pets are from east port'.format(pets_east_port_treated['count(*)'][0]))
```
